### Part A

### Part B

```python
N0 = 1000000 
N1 = 0
results = np.empty([N0, 2], int)

#generate + record results of dice rolling
for i in range(N0):
    dice1 = np.random.randint(1,7)
    dice2 = np.random.randint(1,7)
    results[i] = dice1, dice2

#counting number of times we get double sixes
for c in range(N0):
    e = results[c]
    e_list = str(e.tolist())
    space = ''
    e_string = space.join(e_list)
    if e_string == '[6, 6]':
        N1 += 1

print('# of times double sixes was rolled:', N1)
print('probability of getting double sixes:', N1/N0)
```

### Part C

x' = (ax + c) mod m
- mod m means divide by m and take the remainder

```python
import matplotlib.pyplot as plt

N = 1000
a = 16578 
c = 476877
m = 5.001893

x = 5  #seed
results = []

for i in range(N):      
    x = (a*x+c)%m
    results.append(x)

plt.plot(results,"o")
plt.title('set of numbers generated')
plt.show()

plt.hist(results)
plt.title('histogram of numbers generated')
plt.show()
```
