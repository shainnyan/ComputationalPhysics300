## 8.5 Homework
Write your own program to do the followings
- generate a data with random noise
- choose different polynomial function to fit the data
- discuss which polynomial function yields the best fit and give your reasoning.

```python
import matplotlib.pyplot as plt
import numpy as np

#define the function & paramters for the plot
f = lambda x: x*np.cos(x)
a,b = -5,5
npoints = 40

#generating random data
np.random.seed(0)
x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5

#fitting the curve
for order in range(5, 12, 2):
    fit = np.polyfit(x, y, order, full=True)
    y_p = np.polyval(fit[0], x)
    plt.plot(x,y_p, label='polyfit'+str(order)+str(' error=')+str(fit[1]))

#plot the data
plt.plot(x,y, 'o', label='noisy data')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=8)
plt.title('Fitting a curve to random data')
plt.show()

print('fitted coefficients: ' + str(fit[0]))
```

```python
'''taking a closer look at order 7'''
order = 7

#fitting the curve
fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)

#plot the data
plt.plot(x,y_p, label='polyfit'+str(order)+str(' error=')+str(fit[1]))
plt.plot(x,y, 'o', label='noisy data')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=8)
plt.title('Fitting a curve to random data')
plt.show()
```

The polynomial function with order 7 might be the best fit because it is still a smooth curve like with order 5, but the error is significantly smaller than that of order 5. Any higher order above this and the curve becomes too jagged, because some data points may not actually need to be included in the data.
