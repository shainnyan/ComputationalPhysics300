```python
import numpy as np
import matplotlib.pyplot as plt

#downloading data
data = np.loadtxt("C:/Users/Shainna/Desktop/School/data for phys 300 lec 9 hw.txt")  #501 entries in this

#fourier transform
c = np.fft.fft2(data)

#plot the data
plt.plot(data)
plt.title('original image, 501x501 shape')
plt.savefig('original image (lec 9 hw)')  #477 KB size
plt.show()

plt.plot(np.abs(c))
plt.title('magnitude of coefficients after dft')
plt.show()

#plt.plot(np.abs(c))
#plt.title('magnitude of coefficients after dft (zoomed in)')
#plt.ylim(0, 2)
#plt.show()
```

```python
#attempt to compress data
c2 = np.copy(c)
discard = np.abs(c)<10000
c2[discard] = 0
y_val = np.fft.ifft2(c2).real
y = y_val/501
x = np.linspace(0, 501,len(y))

plt.plot(x, y)
plt.savefig('compressed image (lec 9 hw)') #419 KB size
plt.show()
```
