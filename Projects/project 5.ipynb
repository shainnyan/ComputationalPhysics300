# Deriving linear regression

Linear regression is a form of curve fitting of data where the input data (x) is assumed to have a linear relationship with the output data (y). In other words, the assumed function for x takes some form of $f(x) = mx + b$. 

The basic principle behind curve fitting in general--and so including linear regression--is that the best fit line to a data set is the line that has the least error between the actual data and the numbers produced from the function that is being fit to the data. This can be written as:

$$\text{err} = \sum(y_i - f(x_i))^2$$

where $f(x) = mx + b$. We'll change $m$ to $a$ and so we get:

$$\text{err} = \sum(y_i - (ax_i + b))^2 $$

This method is called the __least squares approach__. To solve this, we must take the derivative with respect to however many unknown factors we have to account for. With the case of linear regression, we need to take the derivative with respect to a and b. This is shown below.

$$\frac{\partial{\text{err}}}{\partial a} = \frac{\partial}{\partial a}{(y_i - ax_i - b)^2} $$
$$0 = -2\sum_i^n x_i (y_i - ax_i -b)$$
and
$$\frac{\partial{\text{err}}}{\partial b} = \frac{\partial}{\partial b}{(y_i - ax_i - b)^2} $$
$$0 = -2\sum_i^n (y_i - ax_i -b)$$

These two equations could be rewritten as 

$$a\sum x_i^2 + b\sum x_i = \sum(x_iy_i)$$
and
$$a\sum x_i + bn = \sum{y_i}$$

n is the number of points.

Using linear algebra, the above equations can also be rewritten like this:

$$
\begin{bmatrix}
       n     & \sum x_i    \\
    \sum x_i & \sum x_i^2  \\
\end{bmatrix}
\begin{bmatrix}
       b \\
       a \\
\end{bmatrix}
=
\begin{bmatrix}
      \sum y_i \\
      \sum (x_iy_i)\\
\end{bmatrix}
$$
and the goal is to solve for the parameters, aka the column with a and b in it.


# Deriving 2nd and 3rd order polynomial regressions

For a polynomial fit, the exact same procedure is followed. The only thing that changes is the function used. A brief run-through of the math for 2nd order and 3rd order polynomials are seen below.

## 2nd order polynomial
$$\text{err} = \sum(y_i - (ax_i^2 + bx_i + c))^2 $$

Take the derivative with respect to each parameter being fitted for:

$$\frac{\partial{\text{err}}}{\partial a} = \frac{\partial}{\partial a}{(y_i - ax_i^2 - bx_i - c)^2} $$
$$0 = -2\sum_i^n x_i^2 (y_i - ax_i^2 - bx_i - c)$$
and
$$\frac{\partial{\text{err}}}{\partial b} = \frac{\partial}{\partial b}{(y_i - ax_i^2 - bx_i - c)^2} $$
$$0 = -2\sum_i^n x_i (y_i - ax_i^2 - bx_i - c)$$
and
$$\frac{\partial{\text{err}}}{\partial c} = \frac{\partial}{\partial c}{(y_i - ax_i^2 - bx_i - c)^2} $$
$$0 = -2\sum_i^n (y_i - ax_i^2 - bx_i - c)$$

Rewrite the equations:

$$a\sum x_i^4 + b\sum x_i^3 + c\sum x_i^2 = \sum(x_i^2 y_i)$$
$$a\sum x_i^3 + b\sum x_i^2 + c\sum x_i = \sum(x_i y_i)$$
$$a\sum x_i^2 + b\sum x_i + cn = \sum y_i$$

Using linear algebra, we get the following matrix:

$$
\begin{bmatrix}
       n       & \sum x_i       & \sum x_i^2     \\
    \sum x_i   & \sum x_i^2     & \sum x_i^3 \\
    \sum x_i^2 & \sum x_i^3     & \sum x_i^4 \\ 
\end{bmatrix}
\begin{bmatrix}
    c   \\
    b   \\
    a   \\
\end{bmatrix}
=
\begin{bmatrix}
    \sum y_i   \\
    \sum (x_iy_i)   \\
    \sum (x_i^2y_i) \\ 
\end{bmatrix}
$$

## 3rd order polynomial
$$\text{err} = \sum(y_i - (ax_i^3 + bx_i^2 + cx_i + d))^2 $$

Take the derivative with respect to each parameter being fitted for:

$$\frac{\partial{\text{err}}}{\partial a} = \frac{\partial}{\partial a}{(y_i - ax_i^3 - bx_i^2 - cx_i - d)^2} $$
$$0 = -2\sum_i^n x_i^3 (y_i - ax_i^3 - bx_i^2 - cx_i - d)$$
and
$$\frac{\partial{\text{err}}}{\partial b} = \frac{\partial}{\partial b}{(y_i - ax_i^3 - bx_i^2 - cx_i - d)^2} $$
$$0 = -2\sum_i^n x_i^2 (y_i - ax_i^3 - bx_i^2 - cx_i - d)$$
and
$$\frac{\partial{\text{err}}}{\partial c} = \frac{\partial}{\partial c}{(y_i - ax_i^3 - bx_i^2 - cx_i - d)^2} $$
$$0 = -2\sum_i^n x_i (y_i - ax_i^3 - bx_i^2 - cx_i - d)$$
and
$$\frac{\partial{\text{err}}}{\partial d} = \frac{\partial}{\partial d}{(y_i - ax_i^3 - bx_i^2 - cx_i - d)^2} $$
$$0 = -2\sum_i^n (y_i - ax_i^3 - bx_i^2 - cx_i - d)$$

Rewrite the equations:

$$a\sum x_i^6 + b\sum x_i^5 + c\sum x_i^4 + d\sum x_i^3 = \sum(x_i^3 y_i)$$
$$a\sum x_i^5 + b\sum x_i^4 + c\sum x_i^3 + d\sum x_i^2 = \sum(x_i^2 y_i)$$
$$a\sum x_i^4 + b\sum x_i^3 + c\sum x_i^2 + d\sum x_i = \sum(x_i y_i)$$
$$a\sum x_i^3 + b\sum x_i^2 + c\sum x_i + dn = \sum y_i$$

Using linear algebra, we get the following matrix:

$$
\begin{bmatrix}
       n       & \sum x_i       & \sum x_i^2     &\sum x_i^3 \\
    \sum x_i   & \sum x_i^2     & \sum x_i^3     &\sum x_i^4 \\
    \sum x_i^2 & \sum x_i^3     & \sum x_i^4     &\sum x_i^5 \\
    \sum x_i^3 & \sum x_i^4     & \sum x_i^5     &\sum x_i^6 
\end{bmatrix}
\begin{bmatrix}
    d   \\
    c   \\
    b   \\
    a 
\end{bmatrix}
=
\begin{bmatrix}
    \sum y_i   \\
    \sum (x_i y_i)   \\
    \sum (x_i^2 y_i) \\
    \sum (x_i^3 y_i) 
\end{bmatrix}
$$


Our code is based on these matrices.

```python
import numpy as np



#function to calculate error
def residual_2(x,y,y_fit):

    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_xy = sum(x*y)
    y_mean = np.mean(y)

    TSS = 0
    RSS = 0

    for i in range(len(y)):

      TSS = TSS + ((y[i]-y_mean)**2)
      RSS = RSS + ((y[i] - y_fit[i])**2)

    R_2 = 1 - (RSS/TSS)

    return R_2
```

```python
import matplotlib.pyplot as plt
import numpy as np
'''linear regression example'''

#define the function
f = lambda x: 3*x+2

#define the paramters for the plot/initialization
a,b = 0,3
npoints = 10
order = 1
x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5
print(x)
#using polyfit from numpy to check against our own code
fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)


#function to perform linear regression (aka 1st order)
def linear_fit(x,y):
    
    
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_xy = sum(x*y)
    
    a = np.array([[npoints, sum_x], [sum_x, sum_x2]])
    b = np.array([sum_y, sum_xy])
    y_int,slope = np.linalg.solve(a, b)
    
    
    return y_int, slope
    
    
#actually performing the linear regression and assigning those results to variables
y_int_lin_fit, m_lin_fit = linear_fit(x,y)

#new f(x) formed
y_own = m_lin_fit*x + y_int_lin_fit  
print("y =", m_lin_fit,"x+",y_int_lin_fit)

R_2 = residual_2(x,y,y_own)
print("R^2 =",R_2)

#plotting the data
plt.plot(x,y, 'o', label='noisy data')
plt.plot(x,f(x), '-', label='actual')
plt.plot(x,y_p, '--', label='np.polyfit')
plt.plot(x,y_own, '--', label='ours')

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend()
plt.show()
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np

'''2nd order polynomial regression'''
plt.figure(figsize=[12, 8])
#define the function
f = lambda x: 1.5*x**2 + -3*x + 2

#define the paramters for the plot
a,b = 0,3
npoints = 100
order = 2

x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5

#using polyfit from numpy to check against our own code
fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)

#function to perform polynomial regression up to order 3
def poly_fit(x,y,order):
    
    
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_x3 = sum(x**3)
    sum_x4 = sum(x**4)
    sum_x5 = sum(x**5)
    sum_x6 = sum(x**6)
    
    sum_xy = sum(x*y)
    sum_x2y = sum((x**2)*y)
    sum_x3y = sum((x**3)*y)
    
    b_0 = 0
    a1 = 0
    a2 = 0
    a3 = 0
    
    if order == 1:
        a = np.array([[npoints, sum_x], [sum_x, sum_x2]])
        b = np.array([sum_y, sum_xy])
        b_0, a1 = np.linalg.solve(a, b)
    elif order == 2:
        a = np.array([[npoints, sum_x,sum_x2], [sum_x, sum_x2,sum_x3],[sum_x2,sum_x3,sum_x4]])
        b = np.array([sum_y, sum_xy,sum_x2y])
        b_0, a1, a2 = np.linalg.solve(a, b)    
    elif order == 3:
        a = np.array([[npoints, sum_x,sum_x2,sum_x3], [sum_x, sum_x2,sum_x3,sum_x4],[sum_x2,sum_x3,sum_x4,sum_x5],[sum_x3,sum_x4,sum_x5,sum_x6]])
        b = np.array([sum_y, sum_xy,sum_x2y,sum_x3y])
        b_0, a1, a2, a3 = np.linalg.solve(a, b)  
    else:
        print("wrong")
                
    return b_0, a1, a2, a3

#print(poly_fit(x,y,2))
#print(len(poly_fit(x,y,3)))

#actually performing the regression and assigning those results to variables
y_int_fit, a1_fit, a2_fit, a3_fit = poly_fit(x,y,order)

#new f(x) formed with data
y_own = a3_fit*x**3 + a2_fit*x**2 + a1_fit*x + y_int_fit
print("y =", a3_fit,"x^3+", a2_fit,"x^2+", a1_fit,"x+",y_int_fit)

R_2 = residual_2(x,y,y_own)
print("R^2 =",R_2)


    
#plotting the data
plt.plot(x,f(x), label='ideal function')
plt.plot(x,y, 'o', label='noisy data')
plt.plot(x,y_p, label='polyfit'+str(order))
plt.plot(x,y_own, label='Ours')

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=12)
plt.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
plt.show()
```

# Example: diffraction equation for constructive interference

This data was taken from an experiment from PHYS 413, or Intermediate Lab I, here at UNLV. A spectrometer was being calibrated to find parameter values to use for a further part of the experiment. To calibrate it, the diffraction equation for constructive interference was used:

$$m\lambda = d (sin \theta_i + sin(\theta_i + \theta_d))$$

where:
- $m$ = order
- $d$ = measurement on diffraction grating in nm
- $\theta_i$ = angle of incidence
- $\theta_d$ = angle of diffraction

$\theta_d$ is the angle measured from the spectrometer. $d$ and $\theta_i$ are our parameters to be found. From the experiment, this is the equation found:

$$m\lambda = 1655(sin 0.04852 + sin(0.04852 + \theta_d))$$

Even though the above equation is a combination of sine functions, let's see how our polynomial regressions fare.

```python
#trying it out with a real life dataset

plt.figure(figsize=[12, 8])

#define the function + all collected user data
f = lambda x: 1655*(np.sin(0.0008468337531) + np.sin(0.0008468337531+x))

m = np.loadtxt("C:/Users/Shainna/Desktop/School/ComputationalPhysics300-master/Projects/order.txt")
wave = np.loadtxt("C:/Users/Shainna/Desktop/School/ComputationalPhysics300-master/Projects/wavelength.txt")
x = np.radians(np.loadtxt("C:/Users/Shainna/Desktop/School/ComputationalPhysics300-master/Projects/angle of diffraction.txt"))
y = m*wave  #multiply together to create a constant
order = 1

def poly_fit(x,y,order):    
    
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_x3 = sum(x**3)
    sum_x4 = sum(x**4)
    sum_x5 = sum(x**5)
    sum_x6 = sum(x**6)
    
    sum_xy = sum(x*y)
    sum_x2y = sum((x**2)*y)
    sum_x3y = sum((x**3)*y)
    
    b_0 = 0
    a1 = 0
    a2 = 0
    a3 = 0
    
    if order == 1:
        a = np.array([[npoints, sum_x], [sum_x, sum_x2]])
        b = np.array([sum_y, sum_xy])
        b_0, a1 = np.linalg.solve(a, b)
    elif order == 2:
        a = np.array([[npoints, sum_x,sum_x2], [sum_x, sum_x2,sum_x3],[sum_x2,sum_x3,sum_x4]])
        b = np.array([sum_y, sum_xy,sum_x2y])
        b_0, a1, a2 = np.linalg.solve(a, b)    
    elif order == 3:
        a = np.array([[npoints, sum_x,sum_x2,sum_x3], [sum_x, sum_x2,sum_x3,sum_x4],[sum_x2,sum_x3,sum_x4,sum_x5],[sum_x3,sum_x4,sum_x5,sum_x6]])
        b = np.array([sum_y, sum_xy,sum_x2y,sum_x3y])
        b_0, a1, a2, a3 = np.linalg.solve(a, b)  
    else:
        print("wrong")
        
    return b_0, a1, a2, a3

#actually performing the regression and assigning those results to variables
y_int_fit, a1_fit, a2_fit, a3_fit = poly_fit(x,y,order)

#new f(x) formed with data
y_own = a3_fit*x**3 + a2_fit*x**2 + a1_fit*x + y_int_fit
print("y =", a3_fit,"x^3+", a2_fit,"x^2+", a1_fit,"x+",y_int_fit)

R_2 = residual_2(x,y,y_own)
print("R^2 =",R_2)


    
#plotting the data
plt.plot(x,f(x), label='ideal function')
plt.plot(x,y, 'o', label='data')
plt.plot(x,y_own, label='our polyfit')

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.legend(fontsize=12)
plt.title('diffraction equation for constructive interference')
plt.show()
```
